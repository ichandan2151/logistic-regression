```python
#logistic regression(classifier) - supervised learning technique
import pandas as pd
df = pd.read_csv('Social_Network_Ads.csv')
df
```

```python
df.info()
```

```python
#purchased column : 0 and 1
#1 - purchased
#0 - not purchased
#input - age and est salary
#output - Purchased
```

```python
#i want to know the exact count of purchased people
df['Purchased'].value_counts()
```

```python
#divide the data into i/p and o/p
x = df.iloc[:,2:4].values
x
```

```python
y = df.iloc[:,4].values
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
print(y.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#scaling or normalisation
#done only for inputs
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
#Apply regrressor/classifier/clusterer

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#8. fitting the model
model.fit(x_train,y_train)
LogisticRegression()
```

```python
#9 predict the variable
y_pred = model.predict(x_test)
y_pred #predictoutput
```

```python
y_test
```

```python
#individul prediction
model.predict([[19,0]])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```
